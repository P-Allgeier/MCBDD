# Factor Analysis

## Overview
Factor Analysis is a statistical method used to identify underlying relationships between variables. It helps reduce a large number of observed variables into a smaller number of latent factors and is also used for things like Image compression.

## Image compression as an example


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FactorAnalysis
from PIL import Image

image = Image.open('resources/snake.jpg').convert('L')
image = np.array(image)

plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.show()
```

here is an 846x883 pixel image (converted to grayscale). We can run this through Python's factor analyzer to then compress the image down to less than half of the factors:

```python
data = image

fa = FactorAnalysis(n_components=350)
factors = fa.fit_transform(data)

reconstructed = np.dot(factors, fa.components_)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original")

plt.subplot(1, 2, 2)
plt.imshow(reconstructed, cmap='gray')
plt.title("Compressed")

plt.show()
```

Now, despite vaporizing more than half of the original image's factors we can still recognize the original image very easily. This is what factor analysis is used for, it can help reduce the amount of "noise" around the actual data we're interested in
